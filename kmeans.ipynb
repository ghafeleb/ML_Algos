```python
# Generating the data
# This particular setting has one distinct cluster and 3 clusters placed close together.
X, y = make_blobs(n_samples=500,
                  n_features=2,
                  centers=4,
                  cluster_std=1,
                  center_box=(-10.0, 10.0),
                  shuffle=True,
                  random_state=1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1])
```
